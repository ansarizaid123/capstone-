```python
def remove_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 -Q1
    lr = Q1 - (1.5 * IQR)
    ur = Q3 + (1.5 * IQR)
    return lr, ur 
```

```python
for i in df_num:
    lr, ur = remove_outliers(df_num[i])
    df_num[i] = np.where(df_num[i]< lr, lr, df_num[i])
    df_num[i] = np.where(df_num[i]> ur, ur, df_num[i])
```

```python
plt.figure(figsize=(20,20))
feature_list = df_num.columns
for i in range(len(feature_list)):
    plt.subplot(5,3, i+1)
    sns.boxplot(y = df_num[feature_list[i]],data = df_num)
    plt.title('Box plot of {}'.format(feature_list[i]))
    plt.tight_layout()
```
